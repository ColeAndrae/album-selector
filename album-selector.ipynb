```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict
import numpy as np

albums_df = pd.read_csv("music.csv")

min_year = 9999
max_year = 0

for row in range(len(albums_df)):
    year = int(albums_df.loc[row, 'release_date'][:4])
    min_year, max_year = min(min_year, year), max(max_year, year)

print(f'from: {min_year} to: {max_year}')
```

```python
years = np.arange(min_year, max_year + 1)
album_count = [0 for _ in range(max_year - min_year + 1)]

for row in range(len(albums_df)):
    year = int(albums_df.loc[row, 'release_date'][:4])
    album_count[year - min_year] += 1

plt.bar(years, album_count)
```

```python
YEAR = '1959'

album_suggestions = albums_df[albums_df['release_date'].str.startswith(YEAR)]

yearly_ranking = np.random.randint(len(album_suggestions))

album = album_suggestions.iloc[yearly_ranking]

global_ranking = album.position

print(f'suggested album: {album.release_name} by {album.artist_name}, ranked #{yearly_ranking + 1} in {YEAR}, and #{global_ranking + 1} of all time.')
```

```python
genres = set()

for row in range(len(albums_df[:10])):
    primary_genres = albums_df.loc[row, 'primary_genres'].split(',')
    for genre in primary_genres:
       genres.add(genre)

genres
```

```python
GENRE = 'Progressive Rock'

album_suggestions = albums_df[albums_df['primary_genres'].str.contains(GENRE)]

yearly_ranking = np.random.randint(len(album_suggestions))

album = album_suggestions.iloc[yearly_ranking]

global_ranking = album.position

print(f'suggested album: {album.release_name} by {album.artist_name}, ranked #{yearly_ranking + 1} in {YEAR}, and #{global_ranking + 1} of all time.')
```
